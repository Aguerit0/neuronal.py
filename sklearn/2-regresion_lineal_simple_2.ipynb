## Análisis de Regresión Lineal Simple: Predicción de Especies de Iris según el Ancho del Pétalo ##

**En este ejercicio de análisis de datos, exploraremos la regresión lineal simple y su aplicación en la predicción de especies de iris utilizando el conjunto de datos 'iris' de Seaborn.** 

**La regresión lineal simple es una técnica estadística que modela la relación entre una variable independiente y una variable dependiente mediante una línea recta. En este caso, utilizaremos el ancho del pétalo ('petal_width') como variable independiente para predecir la especie de iris. Acompáñame mientras exploramos este fascinante método y aplicamos nuestros conocimientos de regresión lineal simple en Python.**

**Importar librerias**

```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
```

**Definir nuestro data frame**

```python
iris = seaborn.load_dataset('iris')
iris.head()
```

**Información de nuestro dataframe**

```python
iris['species'].unique()
```

```python
tipo = {'setosa':0, 'versicolor':1, 'virginica':2}
iris['species'] = iris['species'].replace(tipo)
iris['species'].unique()
```

**Separamos dataframa en datos de prueba y testeo**

```python
#   unidimensionamal
x = iris['petal_width'].values.reshape((-1,1))

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, iris['species'], test_size=0.2)
```

**Graficamos datos de prueba y testeo**

```python
plt.scatter(x_train, y_train, color='black')
plt.scatter(x_test, y_test, color='orange')
plt.show()
```

**Entrenamos nuestro modelo de regresión lineal**

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(x_train, y_train)
```

```python
##  Predicción
predict = modelo.predict(x_test)
predict
```

```python
#   __dict__ : para obtener pendiente e intersección de la predicció
modelo.__dict__
```

**y = a.x + b**

```python
#   score : para comprobar que tal es la predicción
modelo.score(x_test, y_test)*100
```

**Graficamos nuestra predicción con los datos de testeo**

```python
plt.scatter(x_train, y_train, color='black')
plt.scatter(x_test, y_test, color='orange')
plt.plot(x_test, predict, color='red')
plt.show()
```
